# სიჩქარისა და აჩქარების დაშლა რადიალურ და ტრანსვერსალურ მდგენელებად
განვიხილოთ რაიმე სიბრტყე, რომელზედაც არჩეულია ბრუნვის გარკვეული დადებითი მიმართულება (მაგალითად, საათის ისრის მოძრაობის საწინააღმდეგო მიმართულება) და დავუშვათ, რომ წერტილის ტრაექტორია წარმოედგენს ამ სიბრტყეზე მდებარე გარკვეულ $L$ წირს.

 ავიღოთ ზემოხსენებულ სიბრტყეზე პოლარული $Ox$ ღერძი და დავახასიათოთ მოძრავი $M$ წერტილის მდებარეობა $L$ წირზე $\varphi$ და $r$ პოლარული კოორდინატებით. $\varphi$ არის ის დადებითი ან უარყოფითი კუთხე, რომლითაც უნდა მოვაბრუნოთ $Ox$ ღერძი შესაბამისად ბრუნვის დადებითი ან უარყოფითი მიმართლებით, რომ ის დაემთხვეს $OM$ ვექტორს, $r$ წარმოადგენს $OM=$r ვექტორის სიგრძეს.

ცხადია, რომ $M$ წერტილის პოლარული კოორდინატები $\varphi$ და $r$ წარმოადგენს $t$ პარამეტრის (დროის) გარკვეულ ფუნქციებს: $r=r(t),\varphi =\varphi (t)$.

აღვნიშნოთ r$^o$-ით r ვექტორის მგეზავი და გავავლოთ $M$ წერტილზე ორი ერთიმეორის მართობი ღერძი, რომელთაგან, ერთი მათგანის მგეზავია r$^o$, ხოლო მეორის p$^o$, ამასთან p$^o$ ვექტორი მიიღება r$^o$ ვექტორისგან, თუ ამ უკანასკნელს მოვაბრუნებთ $M$ წერტილის ირგვლივ სიბრტყეზე არჩეული ბრუნვის დადებითი მიმართულებით $\frac{\pi}{2}$. ჩვენი მიზანია ვიპოვოთ სიჩქარისა და აჩქარების მდგენელები ამ ორ ღერძზე, რომელთაც პოლარული საკოორდინატო ღერძები ეწოდება.

$M$ წერტილის რადიუს-ვექტორი $r$, ცხადია, ასე შეიძლება წარმოვადგინოთ.<br>
$$\text{r}=r\text{r}^o\text{.}$$
